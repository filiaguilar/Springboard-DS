### Make Data Values Meaningful

```python
age_range = ['18-24','25-34','35-44','45-54','55-64','65+']
age_vals = sorted(df.Age.unique())
#print(age_vals)
df['Age'] = df.Age.replace(age_vals, age_range)
```

```python
gender = ['Male','Female']
gender_vals = sorted(df.Gender.unique())
#print(gender_vals)
df['Gender'] = df.Gender.replace(gender_vals, gender)
```

```python
edu = ['Left school before 16','Left school at 16','Left school at 17','Left school at 18','Some college/university','Professional certificate/diploma','Bachelors','Masters','Doctorate']
edu_vals = sorted(df.Education.unique())
#print(edu_vals)
df['Education'] = df.Education.replace(edu_vals, edu)
```

```python
cntry = ['Australia','Canada','New Zealand','Other','Republic of Ireland','UK','USA']
cntry_vals = [-0.09765,0.24923,-0.46841000000000005,-0.28519,0.21128000000000002,0.9608200000000001,-0.57009]

df['Country'] = df.Country.replace(cntry_vals, cntry)
```

```python
ethn = ['Asian','Black','Mixed-Black/Asian','Mixed-White/Asian','Mixed-White/Black','Other','White']
ethn_vals = [-0.50212,-1.1070200000000001,1.90725,0.12600,-0.22166,0.11440,-0.31685]

df['Ethnicity'] = df.Ethnicity.replace(ethn_vals, ethn)
```

```python
nscore = np.arange(12,61)
nscore_vals = sorted(df.Neuroticism.unique())
#print(nscore_vals)
df['Neuroticism'] = df.Neuroticism.replace(nscore_vals, nscore)
```

```python
escore = [x for x in range(16,60) if x not in [17,57]]
escore_vals = sorted(df.Extraversion.unique())
#print(escore)
#print(escore_vals)
df['Extraversion'] = df.Extraversion.replace(escore_vals, escore)
```

```python
oscore = [x for x in range(24,61) if x not in [25,27]]
oscore_vals = sorted(df.Openness_to_Experience.unique())
#print(oscore)
#print(oscore_vals)
df['Openness_to_Experience'] = df.Openness_to_Experience.replace(oscore_vals, oscore)
```

```python
ascore = [x for x in range(12,61) if x not in [13,14,15,17,19,20,21,22]]
ascore_vals = sorted(df.Agreeableness.unique())
#print(ascore)
#print(ascore_vals)
df['Agreeableness'] = df.Agreeableness.replace(ascore_vals, ascore)
```

```python
cscore = [x for x in range(17,60) if x not in [18,58]]
cscore_vals = sorted(df.Conscientiousness.unique())
#print(cscore)
#print(cscore_vals)
df['Conscientiousness'] = df.Conscientiousness.replace(cscore_vals, cscore)
```

```python
drug_code = ['CL0', 'CL1','CL2','CL3','CL4','CL5','CL6']
use = ['Never used','Used over a decade ago','Used in last decade','Used in last year','Used in last month','Used in last week','Used in last day']

df.iloc[:,13:32] = df.iloc[:,13:32].replace(drug_code, use)
df.head()
```

### Check For Nulls and Duplicate Rows

```python
#Function that calculates percentage of missing data per column
def missingData(dataframe):
    missing = False
    for item in (dataframe.isnull().sum()).iteritems():
        if item[1] > 0:
            print('Missing Data percentage for '+item[0]+' is {:2.2%}'.format((item[1]/dataframe.shape[0])) )
            missing = True
    if not missing:
        print('Found no missing values.')

missingData(df)
```
